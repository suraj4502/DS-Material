```python
ipl = pd.read_csv('Data/ipl-matches.csv')
movies = pd.read_csv('Data/movies.csv')
```

```python
cond = ipl['MatchNumber'] == 'Final'
ipl[cond][['Season','WinningTeam']]
```

```python
# find all the final winners
ipl[ipl['MatchNumber'] == 'Final'][['Season','WinningTeam']]
```

```python
# how many super over finishes have occured
ipl['SuperOver'].unique()
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
# how many matches has RCB won in Banglore
mask1 = ipl['City'] == 'Bangalore'
mask2 = ipl['WinningTeam']== 'Royal Challengers Bangalore'

print(f"THE total number matches won by RCB in Banglore is {ipl[mask1 & mask2].shape[0]}.")
```

```python
# toss winner is match winner in percentage
toss_winner = ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]
total_matches = ipl.shape[0]

np.round((toss_winner/total_matches) * 100,decimals=2)
```
