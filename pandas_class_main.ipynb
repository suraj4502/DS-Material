```python
df.rename(columns={'Division Title' : 'Division',
                  'Division Body_Wins':'Wins',
                  'Division Body_Losses' : 'Losses',
                  'Division Body_Draws': 'Draws'},inplace=True)
```

```python
# Calculate the average age of athletes in each division.
df.groupby('Division')['Bio Data_Age'].mean().sort_values(ascending=False)
```

```python
df.info()
```

```python
df['Losses'].isnull().sum()
```

```python
df['Losses'].unique()
```

```python
df['Wins'] = df['Wins'].replace('Null', pd.NA)
df['Wins'] = pd.to_numeric(df['Wins'], errors='coerce')
```

```python
df['Losses'] = df['Losses'].replace({'Null': pd.NA,'<NA>':pd.NA})
df['Losses'] = pd.to_numeric(df['Losses'], errors='coerce')
```

```python
# Calculate the total number of wins and losses for each division.
df.groupby('Division')[['Wins','Losses']].sum()
```

```python
df['Bio Data_Fighting style'].fillna('unknown',inplace=True)
```

```python
df['Bio Data_Fighting style'].isnull().sum()
```

```python
# Find the most common fighting style among athletes.
temp = df.groupby("Bio Data_Fighting style")['Name'].count().reset_index()
```

```python
result = temp.nlargest(2,'Name').iloc[-1]['Bio Data_Fighting style']
print(f"The most common fighting styles among the atheletes is {result}.")
```
