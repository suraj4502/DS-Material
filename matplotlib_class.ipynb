```python
# line plot
data = {
    'date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    's1': [10, 15, 12, 20, 25, 22, 18, 30, 28, 35],
    's2': [12, 13, 14, 15, 16, 17, 18, 18, 19, 19]
}
time_series = pd.DataFrame(data)
time_series.head()
```

```python
plt.plot(time_series['date'],time_series['s1'],
         marker = "o",
         linestyle = "--",
         c='g', 
         linewidth = 1,
         label='stock1')

plt.plot(time_series['date'],time_series['s2'],
         marker = "^",
         linestyle = "solid",
         c='b', 
         linewidth = 1,
         label='stock2')


plt.title("LIne plot example")
plt.xlabel("Date")
plt.ylabel("Price",fontdict={'size':20},c='r')
plt.legend()
plt.xticks(rotation= 45)
plt.grid(axis= 'y')

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

fig.suptitle("SUBPLOTS")

axes[0, 0].scatter(tips_data['total_bill'],tips_data['tip'])
axes[0, 0].set_title('Total bill vs tips')

axes[0, 1].bar(tips_data['day'],tips_data['total_bill'])
axes[0, 1].set_title('Day Wise Sales')
axes[0, 1].set_ylabel("Sales")

temp = tips_data.groupby('sex')['tip'].mean()
axes[1,0].pie(temp,labels=temp.index)
axes[1, 0].set_title('AVG Tip BY GENDER')

axes[1,1].bar(tips_data['time'],tips_data['tip'])
axes[1, 1].set_title('AVG Tip BY TIME')

# Adjust layout
plt.tight_layout()

plt.show()
```
