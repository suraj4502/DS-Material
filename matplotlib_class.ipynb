```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import seaborn as sns
```

```python
# scatterplot
age = [10,11,12,13,15,16]
height = [60,65,70,75,80,60]
```

```python
plt.scatter(age,height,c = '#FF6969',marker='+')
plt.title("AGE vs Height")
plt.xlabel("Age")
plt.ylabel("Height")
plt.show()
```

```python
sample = pd.read_csv("Data/sample.csv")
sample.head()
```

```python
plt.scatter(sample['weight'],sample['height'],c = '#C23373',marker='o', s = sample['age']*10)
plt.title("WEIGHT vs Height")
plt.xlabel("WEIGHT")
plt.ylabel("Height")
plt.grid(axis='y')
plt.show()
```

```python
# Load data from seaborn
iris_data = sns.load_dataset("iris")
tips_data = sns.load_dataset("tips")
```

```python
# line plot
data = {
    'date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    's1': [10, 15, 12, 20, 25, 22, 18, 30, 28, 35],
    's2': [12, 13, 14, 15, 16, 17, 18, 18, 19, 19]
}
time_series = pd.DataFrame(data)
time_series.head()
```

```python
plt.plot(time_series['date'],time_series['s1'],
         marker = "o",
         linestyle = "--",
         c='g', 
         linewidth = 1,
         label='stock1')

plt.plot(time_series['date'],time_series['s2'],
         marker = "^",
         linestyle = "solid",
         c='b', 
         linewidth = 1,
         label='stock2')


plt.title("LIne plot example")
plt.xlabel("Date")
plt.ylabel("Price",fontdict={'size':20},c='r')
plt.legend()
plt.xticks(rotation= 45)
plt.grid(axis= 'y')
plt.show()
```
