```python
df = pd.read_excel("Data/fighter_details.xlsx")
```

```python
pd.set_option('display.max_columns',None)
```

```python
df.rename(columns ={'Division Title': "Division",
                    'Division Body_Wins': "Wins",
                    'Division Body_Losses' : 'Loses',
                    'Division Body_Draws' : 'Draws',
                    'Bio Data_Status': "Status",
                    'Bio Data_Hometown': "Hometown",
                    'Bio Data_Age': 'Age'},inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df = df.iloc[:,0:46]
```

```python
df.drop(columns=['Records_Wins by Decision'])
```

```python
df['Nickname'] = df['Nickname'].replace({"Null" : "Unknown"})
```

```python
df['Wins'].replace({"Null" : pd.NA},inplace=True)
```

```python
df['Loses'].replace({"Null" : pd.NA},inplace=True)
```

```python
df['Draws'].replace({"Null" : pd.NA},inplace=True)
```

```python
df['Draws'].unique()
```

```python
df['Wins'] = pd.to_numeric(df['Wins'],errors="coerce")
df['Loses'] = pd.to_numeric(df['Loses'],errors="coerce")
df['Draws'] = pd.to_numeric(df['Draws'],errors="coerce")
```

```python
df['Country'] = 66
```

```python
df['Hometown'].fillna('Unknown',inplace=True)
```

```python
def get_country(country_list):
    if len(country_list) > 1:
        return country_list[1]
    else :
        return country_list[0]
```

```python
df['Country'] = df['Hometown'].str.split(',').apply(get_country).str.strip()
# lambda x : x[1] if len(x) > 1 else x[0]
```

```python
df['Total_matches'] = df['Wins'] + df['Loses'] + df["Draws"]
```

```python
df['Score'] =(df['Wins']/df['Total_matches'])*100
```

```python
df['Score_grade'] = df['Score'].apply(lambda x : 'A' if x > 80 else "B" )
```
