```python
# Load data from seaborn
iris_data = sns.load_dataset("iris")
tips_data = sns.load_dataset("tips")
titanic = sns.load_dataset('titanic')
```

```python
tips_data
```

```python
plt.scatter(tips_data['total_bill'], tips_data['tip'],
            c = 'r', marker = '^', s = tips_data['size']*10 )
plt.title("BILL vs TIP")
plt.xlabel("BIll")
plt.ylabel("TIp")
plt.grid(axis = 'y')
plt.show()
```

```python
# line plot
data = {
    'date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    's1': [10, 15, 12, 20, 25, 22, 18, 30, 28, 35],
    's2': [15,25,28,55,98,88,77,55,98,55]
}
time_series = pd.DataFrame(data)
time_series
```

```python
plt.plot(time_series['date'],time_series['s1'],
         linestyle = '--',c = 'g',marker= '^')
plt.plot(time_series['date'],time_series['s2'],
         linestyle = '-',c = 'r',marker= 'o')
plt.xticks(rotation = 45)
plt.title("LIne plot example")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
# PIECHART
categories = ['toys','mobiles','furnitures','accessories']
sales = [100,250,150,50]
```

```python
plt.figure(figsize=(8,8))
plt.pie(sales,labels = categories,colors = ['r','g','b','pink'], 
        autopct ='%1.3f%%', explode =[0,0,0,0.3],shadow=True)
plt.title("PIe chart")
plt.show()
```

```python
# SUBPLOTS
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2,figsize= (10,6))
fig.suptitle("SUBPLOTS")

axes[0, 0].scatter(tips_data['total_bill'],tips_data['tip'])
axes[0, 0].set_title('Total bill vs tips')

axes[0,1].pie(sales,labels = categories,colors = ['r','g','b','pink'])


axes[1,0].hist(tips_data['total_bill'],color = 'r',edgecolor = 'b',bins = 20)

axes[1,1].boxplot(tips_data['tip'],vert=False,patch_artist=True,labels=['Tips'])
plt.tight_layout()
plt.show()
```
