# Seaborn

- *Seaborn is a powerful data visualization library built on top of Matplotlib.* 
- *It is designed to simplify the creation of informative and attractive statistical graphics.* 
- *Seaborn provides a higher-level interface for creating various types of plots, allowing you to create complex visualizations with just a few lines of code.* 
- *It comes with several built-in themes and color palettes that make it easy to create aesthetically pleasing visualizations.*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# loading datasets
tips_data = sns.load_dataset('tips')
gap_data = px.data.gapminder()
iris_data = sns.load_dataset("iris")
```

##  Relational Plots:-

```python
tips_data
```

```python
sns.relplot(data=tips_data,kind='scatter',x='total_bill',y='tip',
            height=5,hue='sex',
            aspect=1.5,col='day',col_wrap=2,palette= 'Reds') #FIGURE LEVEL
plt.show() # FACETTING /FACET GRID
```

```python
sns.relplot(data=tips_data,kind='scatter',x='total_bill',y='tip',
            hue='day',
            style='time',
            size='size',
            height=5,
            aspect=1.5,palette='magma') #FIGURE LEVEL
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=tips_data,x='total_bill',y='tip',hue='day',style='time',size='size') #AXIS LEVEL
plt.title("SCATTER PLOT IN seaborn")
plt.legend(loc='upper left')
plt.show()
```

```python
# SCATTERPLOT
sns.scatterplot(data=tips_data, x='total_bill', y='tip',hue='sex')
#,style='time'
#size='size'
#sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
plt.show()
```

## DISTRIBUTION PLOTS:-

```python
sns.set_style("white")
```

```python
sns.histplot(data=tips_data,x='total_bill',color='#FF6B6B', bins=20)
plt.title("Distribution of Total Bill Amounts")
plt.xlabel("Total Bill Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.displot(data=tips_data, x='total_bill', kind='hist');
# ,hue='sex'
```

```python
# countplot
sns.displot(data=tips_data, x='day', kind='hist') 
```

```python
# kde plot
sns.kdeplot(data=tips_data,x='total_bill')
```

```python
sns.displot(data=tips_data,x='total_bill',kind='kde',hue='day',fill=True,height=10,aspect=2);
```

```python
# Bivariate Histogram
sns.displot(data=tips_data, x='total_bill', y='tip',kind='hist')
```

```python
# BOXPLOTS

sns.boxplot(data=tips_data,y='total_bill')
```

```python
# CATEGORICAL PLOTS:
```

```python
sns.stripplot(data=tips_data,x='day',y='total_bill',hue='sex');
# jitter
# swarmplot
```

```python
sns.catplot(data=tips_data, x='day',y='total_bill',kind='strip');
```

```python
# categorical BOX PLOTS:
sns.boxplot(data=tips_data,x='day',y='total_bill')
# sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
# Violin PLOT ==> box + kde
```

```python
sns.violinplot(data=tips_data,x='day',y='total_bill')
# sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
# ,hue='sex',split=True
```

```python
import numpy as np
sns.barplot(data=tips_data, x='sex', y='total_bill',hue='smoker',estimator=np.min,errorbar=None)
plt.show()
```

```python
# countplot
sns.countplot(data=tips_data,x='sex',hue='day')
```

```python
# faceting using catplot
sns.catplot(data=tips_data, x='sex',y='total_bill',kind='bar',row='time',col='smoker',errorbar=None)
```

```python
# ANOTHER WAY TO PERFORM FACETTING
g = sns.FacetGrid(data=tips_data,col='day',row='time',hue='smoker')
g.map(sns.scatterplot,'sex','total_bill')
g.add_legend()
```
