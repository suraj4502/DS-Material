## SEABORN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# loading datasets
tips_data = sns.load_dataset('tips')
iris_data = sns.load_dataset("iris")
gap_data = px.data.gapminder()
```

# RELATIONAL plots

- Scatter plot
- Line plot

```python
tips_data
```

```python
g = sns.relplot(data=tips_data,
            kind = 'scatter',
            x ='total_bill',
            y = 'tip',
            hue = 'day',
            palette='magma',
            height = 5, 
           aspect=1.6,
               row = 'time',
               col= 'smoker') # figure level chart

g.fig.suptitle("Figure level chart for scatterplot",fontsize = 15,fontweight = 'bold')
plt.show()
```

```python
plt.figure(figsize= (10,6))
sns.scatterplot(data=tips_data,
                x ='total_bill',
                y = 'tip',
                hue = 'day',style = 'time',size = 'size') # axis level
plt.show()
# ,hue='day',style='time',size='size'
```

# Distribution PLots.
- Histogram.
- KDE


```python
sns.kdeplot(data=tips_data, x = 'total_bill', hue= 'sex',
            palette=['r','b'])
```

```python
sns.displot(data=tips_data, x = 'total_bill', hue= 'sex',
            palette=['r','b'],kind='kde',
           col = 'day',col_wrap=2,)
plt.show()  #FIRURE LEVEL
```

```python
sns.histplot(data=tips_data, x = 'total_bill',bins = 20, hue= 'sex',palette='vlag') # axis level
```
