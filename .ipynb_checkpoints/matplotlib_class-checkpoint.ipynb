```python
# line plot
data = {
    'date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    's1': [10, 15, 12, 20, 25, 22, 18, 30, 28, 35],
    's2': [12, 13, 14, 15, 16, 17, 18, 18, 19, 19]
}
time_series = pd.DataFrame(data)
time_series.head()
```

```python
plt.plot(time_series['date'],time_series['s1'],
         marker = "o",
         linestyle = "--",
         c='g', 
         linewidth = 1,
         label='stock1')

plt.plot(time_series['date'],time_series['s2'],
         marker = "^",
         linestyle = "solid",
         c='b', 
         linewidth = 1,
         label='stock2')


plt.title("LIne plot example")
plt.xlabel("Date")
plt.ylabel("Price",fontdict={'size':20},c='r')
plt.legend()
plt.xticks(rotation= 45)
plt.grid(axis= 'y')

plt.show()
```

```python
day_wise_sales = tips_data.groupby('day')['total_bill'].sum().reset_index()
```

```python
day_wise_sales
```

```python
# Day wise sales
plt.bar(day_wise_sales['day'],day_wise_sales['total_bill'], color = 'r')
plt.show()
```

```python
# Gender Wise sales
plt.bar(tips_data.groupby("sex")['total_bill'].sum().index,tips_data.groupby("sex")['total_bill'].sum())
plt.show()
```

```python
# Average tip by gender
tip_by_gender=tips_data.groupby('sex')['tip'].mean()
```

```python
plt.bar(tip_by_gender.index,tip_by_gender)
plt.show()
```

```python
titanic['class'].value_counts()
```

```python
# class wise distribution
class_wise_distribution = titanic['class'].value_counts().reset_index()
class_wise_distribution.rename(columns={'index':'Class','class':'frequency'},inplace=True)
class_wise_distribution
```

```python
plt.bar(class_wise_distribution['Class'],class_wise_distribution['frequency'],
        color = ['r','b','g'],
        label= class_wise_distribution['Class'])
plt.title("Class wise Distrbution of Passengers.")
plt.ylabel("Frequency")
plt.xlabel("Passenger Class")
plt.legend()
plt.show()
```

```python
plt.bar(class_wise_distribution.index,class_wise_distribution,
        color = ['r','b','g'],
        label= class_wise_distribution.index)
plt.title("Class wise Distrbution of Passengers.")
plt.ylabel("Frequency")
plt.xlabel("Passenger Class")
plt.legend()
plt.show()
```

```python
batsman = pd.read_csv('Data/batsman_season_record.csv')
batsman['total'] = batsman['2015'] + batsman['2016'] +batsman['2017']
batsman
```

```python
plt.bar(batsman['batsman'],batsman['total']);
```

```python
plt.bar(batsman['batsman'],batsman['2015'],label = '2015')
plt.bar(batsman['batsman'],batsman['2016'],bottom=batsman['2015'],label='2016')
plt.bar(batsman['batsman'],batsman['2017'],bottom= batsman['2015']+batsman['2016'],label='2017')
plt.legend();
```

```python
# STACKED BAR CHART

plt.bar(batsman['batsman'],batsman['2017'],label = '2017')
plt.bar(batsman['batsman'],batsman['2016'],bottom=batsman['2017'],label='2016')
plt.bar(batsman['batsman'],batsman['2015'],bottom= batsman['2016']+batsman['2017'],label='2015')
plt.title("CRICKET KA CHART")
plt.legend();
```
