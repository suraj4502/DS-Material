```python
def get_country(country_list):
    if len(country_list) > 1:
        return country_list[1]
    else :
        return country_list[0]
```

```python
df['Country'] = df['Hometown'].str.split(',').apply(get_country).str.strip()
# lambda x : x[1] if len(x) > 1 else x[0]
```

```python
df['Total_matches'] = df['Wins'] + df['Loses'] + df["Draws"]
```

```python
df['Score'] =(df['Wins']/df['Total_matches'])*100
```

```python
df['Score_grade'] = df['Score'].apply(lambda x : 'A' if x > 80 else "B" )
```

```python
df['Company'] = df['model'].str.split().apply(lambda x : x[0])
```

```python
df['price'] =df['price'].str.lstrip("₹").str.replace(',',"").astype('float')
```

```python
df['Ram']= df['ram'].str.split(',').apply(lambda z : z[0])
```

```python
df['Storage'] =df['ram'].str.split(',').apply(lambda x : x[1] if len(x)> 1 else 'NOt available')
```

```python
df["Battery Size"]=df['battery'].str.split().apply(lambda x : x[0])
```

```python
df['Fast_charging'] = df['battery'].str.contains('Fast Charging').apply(lambda x : 'Yes' if True else 'NO')
```

```python
df['Display_size']=df['display'].str.split(',').apply(lambda x : x[0])
```

```python
df['Screen refresh rate']=df['display'].str.split(',').apply(lambda x : x[2] if len(x) > 2 else 'Unknown' ).str.split().apply(lambda x : x[0])
```
