```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# loading datasets
tips_data = sns.load_dataset('tips')
gap_data = px.data.gapminder()
iris_data = sns.load_dataset("iris")
```

```python
plot = sns.relplot(data=tips_data,kind='scatter',x='total_bill',y='tip',  # FACETTING /FACET GRID
            height=5,hue='time',
            aspect=1.5,col='day',col_wrap=2,) #FIGURE LEVEL
plot.fig.suptitle("Day wise Analysis")
# .set(title = 'Day wise analysis')
plt.show()
```
