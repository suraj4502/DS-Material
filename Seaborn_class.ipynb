```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# loading datasets
tips_data = sns.load_dataset('tips')
gap_data = px.data.gapminder()
iris_data = sns.load_dataset("iris")
```

```python
temp = gap_data[gap_data['country'] == 'United Kingdom']
temp
```

```python
sns.lineplot(data=temp,x='year',y ='pop')
plt.show()
```

```python
sns.relplot(data=tips_data,kind='scatter',x='total_bill',y='tip',
            height=5,hue='sex',
            aspect=1.5,col='day',col_wrap=2,palette= 'Reds') #FIGURE LEVEL
plt.show() # FACETTING /FACET GRID
```

```python
sns.relplot(data=tips_data,kind='scatter',x='total_bill',y='tip',
            hue='day',
            style='time',
            size='size',
            height=5,
            aspect=1.5,palette='magma') #FIGURE LEVEL
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=tips_data,x='total_bill',y='tip',hue='day',style='time',size='size') #AXIS LEVEL
plt.title("SCATTER PLOT IN seaborn")
plt.legend(loc='upper left')
plt.show()
```
