```python
movies.groupby('Released_Year').get_group('1962').sort_values('Gross',ascending= False)
```

```python
movies.groupby('Released_Year').get_group('1962')['Gross'].idxmax()
```

```python
# find the total earnings by each year
movies.groupby('Released_Year')['Gross'].sum().sort_values(ascending= False)[0:3]
```

```python
# find the top 3 genres by total earning
movies.groupby('Genre')['Gross'].sum().sort_values(ascending= False)[0:3]
```

```python
# find the genre with highest avg IMDB rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending = False)
```
